```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
```

```python
df.head()
```

# Data Cleaning

```python
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace('.','')
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df.head()
```

```python
df.rename(columns= {'Production (Tons)': 'Production(Tons)', 'Production per Person (Kg)' : 'Production_per_Person(Kg)', 'Acreage (Hectare)' : 'Acreage(Hectare)' , 'Yield (Kg / Hectare)': 'Yield(Kg/Hectare)' }, inplace =True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[df['Acreage(Hectare)'].isnull()]
```

```python
df = df.dropna().reset_index()
```

```python
df.head()
```

```python
df.drop(["index","Unnamed: 0"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Production(Tons)"] = df["Production(Tons)"].astype(float)
df["Production_per_Person(Kg)"] = df["Production_per_Person(Kg)"].astype(float)
df["Acreage(Hectare)"] = df["Acreage(Hectare)"].astype(float)
df["Yield(Kg/Hectare)"] = df["Yield(Kg/Hectare)"].astype(float)
```

```python
df.dtypes
```

```python
df.nunique()
```

# Analysis for Continent

```python
df_cont = df.groupby("Continent").sum()
```

```python
df_cont
```

## Continent with maximum Production

```python
df_cont["Production(Tons)"].sort_values(ascending = False).plot(kind = "bar")
```

## Do number of countries in a continent affect sugarcane production ?

```python
df_cont["number_of_countries"] = df.groupby("Continent").count()["Country"]
cont_names = df_cont.index.to_list()
sns.lineplot(data = df_cont, x = "number_of_countries", y = "Production(Tons)")
plt.xticks(df_cont["number_of_countries"],cont_names, rotation = 90)
plt.show()
```

## Do continent with highest land produce more sugarcane ?

```python
ax = sns.lineplot(data = df_cont, x = "Acreage(Hectare)", y = "Production(Tons)")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
plt.show()
```
